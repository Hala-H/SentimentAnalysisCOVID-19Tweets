# Read JSON file

To avoid memory errors, we split the json file containing all tweets into 4 files using this command in bash: 
split -l 250000 'output.json'
The result was 4 files with 250,000 rows each.
Next, we started reading each file into a dataframe and then combined them all in one dataframe.

```python
import pandas as pd
import json
```

```python
import json

tweetsList = []
print("Started Reading JSON file which contains multiple JSON document")

with open('xaa.json', encoding = 'utf-8') as f:
    for jsonObj in f:
        tweetsDict = json.loads(jsonObj)
        tweetsList.append(tweetsDict)
```

```python
df_1 = pd.DataFrame(tweetsList)
```
